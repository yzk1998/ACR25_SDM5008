```python
import numpy as np

A = np.array([[1,-2,4],[1,-1,1],[1,0,0],[1,1,1]])
B = np.array([[1,2,3],[1,2,3],[1,2,3],[1,2,3]])
b = np.array([[0],[0],[1],[0]])
print(f"A:{A}\nB:{B}\nb:{b}")

print(f"The second row of A: {A[1, :]}")
print(f"The third column of B: {B[:,2]}")

print(f"A+B: {A+B}")

print(f"B append to A:{np.append(A,B,axis=1)}")

#矩阵相乘
print(f"AT*B: {A.T@B}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建角度数组（0到2π）
theta = np.linspace(0, 2*np.pi, 100)

# 计算单位圆上的点
x = np.cos(theta)
y = np.sin(theta)

# 创建图形和坐标轴
plt.figure(figsize=(8, 8))
ax = plt.subplot(111)

# 绘制单位圆
ax.plot(x, y, 'b-', linewidth=2, label='unit circle')

# 设置坐标轴
ax.axhline(y=0, color='k', linewidth=0.5)  # 水平轴
ax.axvline(x=0, color='k', linewidth=0.5)  # 垂直轴
ax.set_aspect('equal')  # 确保比例相等，圆看起来是正圆

# 添加网格
ax.grid(True, linestyle='--', alpha=0.7)

# 设置坐标轴范围
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)

# 添加标签和标题
ax.set_title('unit circle', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)

InnerP_theta = np.linspace(0, 2*np.pi, 11)

InnerP_x = np.cos(InnerP_theta)
InnerP_y = np.sin(InnerP_theta)

ax.plot(InnerP_x, InnerP_y, '+', markersize=15)

# 显示图形
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Given parameters
P = np.array([[4, 1], [1, 4]])
xc = np.array([1, 2])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(P)
lambda1, lambda2 = eigenvalues
v1, v2 = eigenvectors[:, 0], eigenvectors[:, 1]

# Semi-axis lengths
a = np.sqrt(lambda1)  # major axis length
b = np.sqrt(lambda2)  # minor axis length

# Rotation angle (angle of major axis with x-axis in degrees)
angle = np.degrees(np.arctan2(v1[1], v1[0]))

# Create ellipse patch
ellipse = Ellipse(xy=xc, width=2*a, height=2*b, angle=angle,
                  edgecolor='b', facecolor='none', linewidth=2)

# Plot
fig, ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_xlim(-2, 4)
ax.set_ylim(-1, 5)
ax.set_aspect('equal')
ax.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
